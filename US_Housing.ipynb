```python
u=np.mean(x,axis=0)
```

```python
std=np.std(x,axis=0)
```

```python
x_train=(x_train-u)/std
```

```python
x_test=(x_test-u)/std
```

```python
a=np.mean(y,axis=0)
```

```python
b=np.std(y,axis=0)
```

```python
y_test=(y_test-a)/b
```

```python
y_train=(y_train-a)/b
```

```python
x_train=np.array(x_train)
y_train=np.array(y_train)
x_test=np.array(x_test)
x_train=np.array(x_train)
```

```python
plt.style.use('seaborn')
plt.scatter(x_train[:,0],y_train)
plt.show()
```

```python
ones=np.ones((x_train.shape[0],1))
x_train=np.hstack((ones,x_train))
```

```python
print(x_train[:5,])
```

```python
def hypothesis(x,theta):
  return np.dot(x,theta)
```

```python
def cost(x,y,theta):
  m=x.shape[0]
  total_error=0
  y_p=hypothesis(x,theta)
  total_error=np.sum((y_p-y)**2)
  return total_error/m
```

```python
def gradient(X, Y, theta):

    y_p = hypothesis(X, theta)

    grad = np.dot(X.T, (y_p - Y))

    return grad/X.shape[0]
```

```python
def gradient_descent(X, Y, learning_rate = 0.05, max_steps = 1000):

    m, n = X.shape
    theta = np.zeros((n, ))
    cost_epoch = []

    for i in range(max_steps):

        grad = gradient(X, Y, theta)

        e = cost(X, Y, theta)
        cost_epoch.append(e)

        theta = theta - learning_rate*grad

    return theta, cost_epoch
```

```python
theta, cost_epoch = gradient_descent(x_train, y_train)
```

```python
ones=np.ones((x_test.shape[0],1))
x_test=np.hstack((ones,x_test))
```

```python
plt.plot(cost_epoch)
```

```python
y_pred_test=[]
for i in range(x_test.shape[0]):
    pred=hypothesis(x_test[i],theta)
    y_pred_test.append(pred)
y_pred_test=np.array(y_pred_test)
```

```python
from sklearn import metrics
print('MAE',metrics.mean_absolute_error(y_test,y_pred_test))
print('MSE',metrics.mean_squared_error(y_test,y_pred_test))
print('RMSE',np.sqrt(metrics.mean_squared_error(y_test,y_pred_test)))
print('Accuracy',metrics.explained_variance_score(y_test,y_pred_test))
```
